### Dropping

```python
X = df.drop(columns=target_var, axis=1).copy()
y = df[target_var]
```

```python
X.columns
```

```python
feats = ['danceability_mean','energy_mean',
    'loudness_mean','mode_mean',
    'speechiness_mean',       'acousticness_mean','liveness_mean',
    'valence_mean','tempo_mean',
    'instrumentalness_mean', 'Cminor_count', 'CMajor_count', 'C#/Dbminor_count', 'C#/DbMajor_count',
       'DMajor_count', 'D#_EbMajor_count', 'Eminor_count', 'EMajor_count',
       'Fminor_count', 'FMajor_count', 'F#/Gbminor_count', 'GMajor_count',
       'G#/Abminor_count', 'G#/AbMajor_count', 'Aminor_count', 'AMajor_count',
       'A#/Bbminor_count', 'BMajor_count', 'Dminor_count', 'D#_Ebminor_count',
       'Gminor_count', 'A#/BbMajor_count', 'F#/GbMajor_count', 'Bminor_count']
```

```python
feats_remove = []

for i in X.columns:
    if i not in feats:
        feats_remove.append(i)
        
feats_remove
```

```python
X.drop(columns=feats_remove, axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

## Preprocessor

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler
from sklearn.compose import make_column_transformer

num_feats = X.select_dtypes("number").columns

imputer = SimpleImputer(strategy="median")

scaler = MinMaxScaler()

num_trans = make_pipeline(imputer, scaler)

preprocessor = make_column_transformer((num_trans, num_feats))
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

ml_models = [RandomForestClassifier(n_jobs=-1), 
             SVC(), LogisticRegression(max_iter=1000, n_jobs=-1, multi_class="multinomial"), 
             GradientBoostingClassifier(),
             KNeighborsClassifier(), AdaBoostClassifier()
            ]

for model in ml_models:
    pipe = make_pipeline(preprocessor, model)
    grid = GridSearchCV(estimator=pipe, param_grid={}, cv=5, scoring="accuracy")
    grid.fit(X_train, y_train)
    score = grid.best_score_
    
    print(f"Classifier: {model} scored {score}\n")  
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
pipe = make_pipeline(preprocessor, RandomForestClassifier())
grid = GridSearchCV(estimator=pipe, param_grid={}, cv=5, scoring="accuracy")
grid.fit(X_train, y_train)
```

```python
ConfusionMatrixDisplay.from_estimator(grid, X_train, y_train)
```

```python
ConfusionMatrixDisplay.from_estimator(grid, X_test, y_test)
```
